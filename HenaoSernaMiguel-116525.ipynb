```python
#2
#Leer dos números enteros y determinar si la magnitud de la diferencia entre 
#los dos es un número primo.
def leer_entero(): #se define una funcion para que el usuario ingrese los
                   #numeros 
    '''funcion para leer numeros enteros'''
    while True:
        try:
            x = int(input('ingrese un numero\n'))
            break
        except:
            print('solo ingrese valores numericos')
            
    return x

#para este programa se define como diferencia entre dos numeros, la distancia 
#que hay entre ellos en la recta de los reales por lo tanto se aceptan numeros
#negativos

x = leer_entero() #se piden los numeros al usuario
y = leer_entero()

DM = abs(x-y) #magnitud de la diferencia entre los dos numeeros    

cprimo = 0 #con esta variable se evaluara si DM es primo o no lo es

for i in range (1,DM+1): #se emplea un ciclo for para evaluar si la diferencia 
    if DM % i == 0:      #tiene mas de dos o solo dos divisores exactos
        cprimo += 1 
        
if cprimo == 2: #si el numero es primo significa que solo tiene dos divisores
                #que da como resultados residuo igual a cero el 1 y el mismo
    print(f'la magnitud de la diferencia ente {x} y {y} es igual a {DM} y {DM}\
 es un numero primo')

else: print(f'la magnitud de la diferencia ente {x} y {y} es igual a {DM} y\
 {DM} no es numero primo')
```

```python
#5
#Escribir un programa que reciba una cadena de texto y reporte:
#a) cuántas letras vocales en mayúscula se entraron;
#b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
#c) cuántos dígitos se entraron,
#d) cuántos espacios se entraron;
#e) cuántas palabras reservadas se entraron.
import keyword #se importa la libreria de las palabras reservadas

cad = str(input('digite una cadena de de texto\n')) #se pide al usuario que
                                     #entre la cadena
Cvocalesmay = 0 #contador de volcales    
for i in cad: #iterando por cada elemento de la cadena
#el siguiente condicional solo tiene en cuenta vocales mayusculas con o sin
#tilde    
    if i == 'A' or i == 'E' or i == 'I' or i == 'O' or i == 'U' or i == 'Á' or\
       i == 'É' or i == 'Í' or i == 'Ó' or i == 'Ú':
        Cvocalesmay += 1 #si el elemento evaluado es una vocal el contador 
                         #aumentara en uno
Cdigitos = 0 #contador de digitos   
for i in cad: #iterando por cada elemento de la cadena
    if i.isdigit() == True: #con el metodo isdigit se evalua si cada elemento 
                            #elemento de la cadena es o no es un digito
        Cdigitos += 1 #si el elemento es un digito el contador aumentara en uno      

Ctildes = 0 #contador de tildes         
for i in cad.upper(): #se uliza el metodo .upper() para convertir toda la
                      #en mayuscula y asi solo evaluar tildes en mayuscula
   if i == 'Á' or i == 'É' or i == 'Í' or i == 'Ó' or i == 'Ú':
       Ctildes += 1 #si la letra tiene tilde el contador aumentara en uno
    
Cespacios = 0 #contador de espacios
for i in cad: #se itera por cada elemto de la cadena
    if i == ' ': #si el elemento es un espacio
        Cespacios += 1 #si el elemento es un espacio el contador aumentara en
                       #uno

L = cad.split() #se utiliza el metodo split para volver la cadena de texto una
                #una lista de palabras
key = keyword.kwlist #se crea una lista de palabras reservadas

Ckey = 0 #contador de palabras reservadas
for i in L: #iterando por la lista de palabras de la cadena 
    for r in key: #iternado por la lista de las palabras reservadas
        if r == i: #se compara si algun elemento de estas dos listas es igual
            Ckey += 1 #si una palabra de la cadena de texto es un palabra
                      #reservada entonces el contador aumentara en uno
    
print(f'''hay:
{Cvocalesmay} vocales en mayuscula
{Ctildes} letras con tilde
{Cespacios} espacios
{Cdigitos} digitos
{Ckey} palabras reservadas''') #se imprimer los resultados
```

```python
#6
#Leer una cadena de texto y organizar alfabéticamente cada una de las letras 
#que la componen, repitiendo cada una tantas veces como se encuentra. Por 
#ejemplo, la cadena ‘tarea importante’ será ‘aaaeeimnoprrttt’. (Note que no se 
#incluyen los espacios).

cad = str(input('Escriba una cadena de texto\n')) #se pide al usuario ingresar
                                                  #la cadena de texto
cad1 = cad #se utiliza esta variable para no cambiar la variable original
cad1 = cad1.replace(' ', '') #se eliminan los espacios de la cadena de texto
cad_ordenada = str('') #se crea una cadena vacia en la que se adicionara la
                       #cadena ordenada
while True: #ciclo haga hasta
    menor = cad1[0] #en esta variable se almacenara el caracter menor dentro de
                    #la cadena 
    for i in cad1:    #con este ciclo se compara cada caracter para saber cual 
        if i <= menor:#es menor dentro de la cadena
            menor = i
                
    Cmenor = 0 #contador de cuantas veces esta el caracter anterior dentro de
               #de la cadena  
    for i in cad1:     #con este ciclo se cuenta las veces que el caracter esta
        if menor == i: #dentro de la cadena de texto
            Cmenor += 1
    cad_ordenada += Cmenor*menor #se adiciona estos caracteres encontrados en 
                                 #la cadena ordenada
    cad1 = cad1.replace(menor, '') #se elemina los caracteres encontrados de la            
                                   #cadena para empezar de nuevo el proceso
    if cad1 == '': #condicion del ciclo haga hasta 
        break
        
print(f'''"{cad}" odenada alfabeticamente es:
{cad_ordenada}''') #se imprimen los resultados 
```

```python
#9
#Escribir un programa que calcule términos de la sucesión Un+1 = 3 Un + 1 si 
#Un es impar y U n+1 = Un / 2 si Un es par. El programa tiene que pedir el 
#término U0 y el número de términos a calcular.

import numpy as np #se importa la libreria
def leer_entero(y): #funcionn para leer cualquier numero
    '''funcion para leer numeros enteros'''
    while True:
        try:
            x = int(input(f'{y}\n'))
            break
        except:
            print('solo ingrese valores numericos enteros')
            
    return x

Uo = leer_entero('Digame el valor de U(0):  ') #se llama la funcion

terminos = leer_entero('Digame cuantos terminos tiene:  ') #se vuelve a llamar
                                                           #la funcion
L = np.zeros(terminos) #se crea una lista de zeros del numero de terminos
                       #indicado
L[0] = Uo #el primer valor de la lista es u(0)

for i in range(len(L)-1): #iternando por los indices de la lista
    if L[i]%2 == 0: #si el numero es par entonces el calculo es el siguiente
        L[i+1] = L[i]//2
        
    else: #si el numero no es par el calculo es el siguiente
        L[i+1] = 3*L[i] +1

print(f'Los terminos de la sucesion son:\n{L}') #se imprime los resultados
```

```python
#10
#Leer una matriz e imprimir el vector que aparece cuando la matriz se desenreda 
#en forma de espiral que gira en el sentido horario a partir de la esquina 
#inferior derecha.
def leer_matriz():
    '''funcion para leer una matriz de enteros'''
    def leer_entero(y): #funcionn para leer cualquier numero
        '''funcion para leer numeros enteros'''
        while True:
            try:
                x = int(input(f'{y}\n'))
                break
            except:
                print('solo ingrese valores numericos enteros')
                
        return x
    x = leer_entero('Ingrese el numero de filas de la matriz')  #numero de 
                                                                #filas de la 
                                                                #matriz
    y = leer_entero('Ingrese el numero de columnas de la matriz')  #numero de 
                                                                   #columnas de 
                                                                   #la matriz
    MATRIZ = [[0 for c in range(y)] for f in range(x)] #secrea una matriz de
                                                       #ceros del tamaño 
                                                       #ingresado anteriormente 
    
    for i in range(x):      #se itera por cada elemento de la matriz para
        for r in range(y):  #pedirle al usuario entrar el dato correspondiente
            MATRIZ[i][r] = leer_entero(f'ingrese el valor ({i+1}, {r+1}) de \
su matriz')            
    return(MATRIZ) #la funcion una matriz
    
M = leer_matriz() #se llama la funcion leer_matriz para ingresar la matriz
M_copia = M.copy()  #copio la matriz
L = [] #lista cuando la matriz se desenredo

while True:
    for i in range (-1,-len(M_copia[0])-1,-1): #extrae toda la fila de abajo al reves
        L.append(M_copia[-1][i])      
    for i in range (-2,-len(M_copia),-1): #extrae la columna 0 al reves
        L.append(M_copia[i][0])
    for i in range (0,len(M_copia[0])): #extrae la primera fila
        L.append(M_copia[0][i]) 
    for i in range (1,len(M_copia)-1): #extrae toda la ultima columna en el orden 
        L.append(M_copia[i][-1])       #necesario
    
    B=[]                            #extrae una matriz para volver a empezar a 
    for i in range (1,len(M_copia)-1):    #extraer cosas (matriz interna)
        B.append(M_copia[i][1:len(M_copia[i])-1])    
    
    M_copia = B                            
 
    if len(M_copia) == 0: break
    
print(f'''el vector solicitado es:{L}''')
```
